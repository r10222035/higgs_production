# Superviesed learning with Particle Transformer

## With photon

```python
i = 1
for seed in range(123, 1100, 100):
    config_path = f'config_files/supervised_ParT_config_{i:02}.json'
    with open(config_path, 'w') as f:
        data = {}

        data['npy_paths'] = ['../Sample/data/supervised/pre-processing/']
        data['seed'] = seed
        data['model_name'] = f'ParT_w_a_{i:02}'
        data['model_structure'] = 'ParT_Light'
        data['sample_type'] = f'pure, w/ photon'
        data['remove_decay_products'] = False

        data['training_method'] = 'supervised'
        data['n_train'] = 100000
        data['n_val'] = 25000
        data['n_test'] = 25000

        json.dump(data, f, indent=4)

    cmd = f'python train_ParT.py {config_path}'
    print(cmd)
    i += 1
```
